# Normalization of Data 

```python
testData ['exam_1'] = (testData['exam_1'] - testData['exam_1'].min())/ (testData['exam_1'].max()-testData['exam_1'].min())
```

```python
testData ['exam_2'] = (testData['exam_2'] - testData['exam_2'].min()) / (testData['exam_2'].max()-testData['exam_2'].min())
```

# Cost Function

```python
length = len(testData)
```

```python
x0 = np.ones((length))
allX = np.array([ x0, testData['exam_1'], testData['exam_2'] ]).T
```

```python
allW = np.array([0, 1, 2]) #assign random values for w0, w1, w2
```

```python
def costFunction( Y, h ):   
    return (-Y * np.log(h) - (1 - Y) * np.log(1 - h)).mean()
```

# Sigmaoid Function

```python
def sigmaoid (z):
    res = 1.0 / (1+np.exp(-z))
    return res
```

# Predictions

```python
def probabilityOfAdmitted(X, W):
    val = np.dot(X, W)
    p = sigmaoid(val)
   
    return p
```

```python
def predict(X, W):
    val = np.dot(X, W)
    p = sigmaoid(val)
    
    p [p>=0.5] = 1
    p [p<0.5]  = 0
    
    return p
```

# Application of Gradient Descent

```python
def gradientDescent(X, Y, W, alpha = 0.1, iteration = 10000):
    
    costH= []
    for i in range(0,iteration+1):
        x = np.dot(X, W)
        h = sigmaoid(x)
        g = np.dot(X.T, (h - testData['admitted'])) / len(testData['admitted'])
        W = W - alpha * g
        
        cost = costFunction(testData['admitted'], h)
        costH.append(cost)
    
        if i % 1000 == 0: 
            print("Iteration %d: "%i)
            print(cost)
            print()
    
    return W, costH
```

```python
updated_W, costs = gradientDescent(allX, testData['admitted'], allW)
```

```python
plt.plot(costs)
plt.ylabel('J: Cost')
plt.xlabel('Iterations')

plt.show()
```

```python
print(predict(allX, updated_W))
```
