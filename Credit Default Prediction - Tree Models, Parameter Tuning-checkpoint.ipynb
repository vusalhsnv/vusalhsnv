```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Credit Default Dataset.csv')
```

```python
df
```

```python
df.info()
```

```python
df['credit_line_utilization'] = df['credit_line_utilization'].str.replace(',','').astype(float)
```

```python
X = df.drop(columns='defaulted_on_loan', axis=1)
```

```python
y = df['defaulted_on_loan']
```

### Outlier cleaning

```python
X.isna().sum()
```

```python
X.dropna()
```

```python
X_zscore = (X-X.mean()) / X.std()
```

```python
for col in X.columns:
    zscore_col = col + '_zscore'
    X[zscore_col] = X_zscore[col]
```

```python
def replace_outliers_with_nan():
    for col in X.columns[:10]:
        zscore_col = col + '_zscore' 
        X.loc[((X[zscore_col] > 3) | (X[zscore_col] < -3)), col] = np.nan
```

```python
replace_outliers_with_nan()
```

```python
X.columns[10:20]
```

```python
for col in X.columns[10:20]:
    X.drop(columns=col, axis=1, inplace=True)
    
```

```python
X
```

```python
X.isna().sum()
```

```python
df['defaulted_on_loan'].value_counts(normalize=True)
```

## Train_Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, shuffle =True, random_state = 42)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

## Missing Value Imputation

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy= 'median')
```

```python
columns = X.columns
```

```python
X_train[columns] = imputer.fit_transform(X_train[columns])
```

```python
X_train.isna().sum()
```

```python
X_test[columns] = imputer.fit_transform(X_test[columns])
```

```python
X_test.isna().sum()
```

### Scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_test
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion = 'entropy')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

### Evaluation

```python
from sklearn.metrics import plot_confusion_matrix, accuracy_score, f1_score, classification_report
```

```python
plot_confusion_matrix(model, X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
f1_score(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
