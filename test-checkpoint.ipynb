```python
def costFun(allX,y,W):
    pred = X.dot(W)
    res = np.sum(((pred - y)**2)/(2*len(y)))
    
    return res
```

```python
def gradientDescent(X,Y,W,alpha = 0.001, iterations =10000 ):
    
    costH = []
    
    
    
    length = len(Y)
    for i in range(0,iterations):
        if i % 1000 == 0:
            print('iteration %d: '%i)
            print(costFun(X,Y,W))
            
        h = X.dot(W)
        diff = h - Y
        g =  X.T.dot(diff) / length
        
        W = W - alpha * g
        
        cost = costFun(X,Y,W)
        
        costH.append(cost)
    return W, costH
```
