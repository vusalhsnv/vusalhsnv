```python
import numpy as np

def ecdis(X,Y):
    return np.linalg.norm(X-Y)
```

```python

centroids

K =3
```

```python
historyX1 = []
historyY1 = []

historyX2 = []
historyY2 = []

historyX3 = []
historyY3 = []


for i in range(30):
    
    scaleCentroid = []
    
    historyX1.append(centroids[0][0])
    historyY1.append(centroids[0][1])
    
    historyX2.append(centroids[1][0])
    historyY2.append(centroids[1][1])
    
    historyX3.append(centroids[2][0])
    historyY3.append(centroids[2][1])
    
    
    for j in range(K):
        scaleCentroid.append([])
        
    for coordinates in df.values:
        distance = []
        for c in centroids:
            distance.append(ecdis(coordinates,c))
            
        dmin = np.argmin(distance)
        scaleCentroid[dmin].append(coordinates)
        
    for k in range(K):
        centroids[k] = np.average(scaleCentroid[k], axis = 0)
```
