```python
def scorerfmValues(temp_data,frequency,recency,monetary):
    if len(temp_data[temp_data[recency] < 0]) == 0:
        temp_data[recency] *= -1  
    
    columns = [frequency,recency,monetary]
    for i in columns:
        q1 = temp_data[i].quantile(.25)
        q2 = temp_data[i].quantile(.50)
        q3 = temp_data[i].quantile(.75)
        
        temp_data[i + '_score'] = 0
        temp_data.loc[temp_data[i]<q1, i + '_score'] = 1
        temp_data.loc[(temp_data[i] >= q1)&(temp_data[i] <= q2), i+ '_score'] = 2
        temp_data.loc[(temp_data[i] > q2)&(temp_data[i] <= q3), i+ '_score'] = 3
        temp_data.loc[temp_data[i] >= q3, i + '_score'] = 4
        
    temp_data[recency] *= -1 
    return temp_data
    
rfm_final = scorerfmValues(rfm_data2,'frequency',recency = 'recency', monetary = 'amount' )
```

```python
def labelData(score):
    if score == 1:
        return 'bad'
    elif score == 2:
        return 'not bad'
    elif score == 3:
        return 'good'
    else:
        return 'best'
```
