```python
import pandas as pd
```

```python
df = pd.read_csv('Datas/california_housing_train.csv')
```

```python
df
```

```python
df.drop(['longitude', 'latitude'], inplace=True, axis=1)
```

```python
X = df.drop('median_house_value', axis=1)
y = df.median_house_value
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
svr = SVR(kernel='rbf')
```

```python
svr.fit(X, y)
```

```python
df_test = pd.read_csv('Datas/california_housing_test.csv')
```

```python
df_test
```

```python
df_test.drop(['longitude', 'latitude'], inplace=True, axis=1)
```

```python
df_test
```

```python
X_test = df.drop('median_house_value', axis=1)
y_test = df.median_house_value
```

```python
pred = svr.predict(X_test)
```

```python
mean_squared_error(pred, y_test, squared=False)
```

```python
mean_absolute_error(pred, y_test)
```

```python
r2_score(pred, y_test)
```
