### Train_Test_Split

```python
# Creating train and test data

X = df.drop(columns=['Life expectancy'])
y = df[['Life expectancy']]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train = scaler.fit_transform(y_train)
y_test = scaler.transform(y_test)
```

```python
X_train
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

reg_model = LinearRegression(fit_intercept = True)
```

```python
reg_model.fit(X_train, y_train)
```

```python
reg_model.score(X_test, y_test)
```

```python
X_train.shape
```

```python
print(reg_model.coef_)
```

```python
print(reg_model.intercept_)
```

```python
y_pred = reg_model.predict(X_test)
```

```python
plt.plot(y_test, y_pred, "^", color='r')
plt.xlabel('Model Prediction')
plt.ylabel('True Values')
```

```python
y_pred_org  =scaler.inverse_transform(y_pred)
y_test_orig  =scaler.inverse_transform(y_test)
```

```python
plt.plot(y_test_orig, y_pred_org, "^", color='r')
plt.xlabel('Model Prediction')
plt.ylabel('True Values')
```

```python
y_pred_org
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
RMSE = np.sqrt(mean_squared_error(y_test_orig, y_pred_org))
MSE = mean_squared_error(y_test_orig, y_pred_org)
MAE = mean_absolute_error(y_test_orig, y_pred_org)
r2 = r2_score(y_test_orig, y_pred_org)

print('RMSE =', RMSE, '\nMSE =', MSE, '\nMAE =', MAE, '\nR2 =', r2)
```

# Support Vector Regressor

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

y_train = scaler.fit_transform(y_train)
y_test = scaler.transform(y_test)
```

```python
from sklearn.svm import SVR

regressor = SVR()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
regressor.score(X_test, y_test)
```

```python
plt.plot(y_test, y_pred, "^", color='r')
plt.xlabel('Model Prediction')
plt.ylabel('True Values')
```

```python
y_pred_org  =scaler.inverse_transform(y_pred.reshape(-1,1))
y_test_orig  =scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
plt.plot(y_test_orig, y_pred_org, "^", color='r')
plt.xlabel('Model Prediction')
plt.ylabel('True Values')
```

```python
RMSE = np.sqrt(mean_squared_error(y_test_orig, y_pred_org))
MSE = mean_squared_error(y_test_orig, y_pred_org)
MAE = mean_absolute_error(y_test_orig, y_pred_org)
r2 = r2_score(y_test_orig, y_pred_org)

print('RMSE =', RMSE, '\nMSE =', MSE, '\nMAE =', MAE, '\nR2 =', r2)
```
